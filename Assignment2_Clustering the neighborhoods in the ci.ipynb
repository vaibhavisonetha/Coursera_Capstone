# importing libraries

```python
import pandas as pd
import numpy as np
import requests
```

# Scrapping data from wikipedia

```python
w = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wpage = requests.get(w)
```

# Creating dataframe and removing not assigned 

```python
wraw = pd.read_html(wpage.content, header = 0)[0]
df = wraw[wraw.Neighborhood != 'Not assigned']
df.reset_index(inplace = True)
df.head()
```
